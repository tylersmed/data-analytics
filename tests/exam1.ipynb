```python
# import and read the csv dataset
import numpy as np
import pandas as pd
# Flights from Houston,TX data set
flights = pd.read_csv("https://github.com/kiat/Elements-of-Data-Analytics/blob/main/datasets/airline/HoustonAirline.csv?raw=true") 
flights
```

```python
# Question 4
"""We can identify each specific airplane using the "TailNum" column of this dataset. 
Develop a Python Pandas implementation that can find the most frequently used airplane within this dataset."""

# find how many times each tail number was used, find the max and save that ID
tailID = flights['TailNum'].value_counts().idxmax()

print('Most frequent tail ID:', tailID)
```

```python
# Question 5
"""Over the entire dataset, which day of the week has the highest flight traffic (i.e. the busiest day of the week)?
Include canceled flights in your calculation."""

# find how many flights occured on every day, find the max, and save the ID for that day of week
maxDOW = flights['DayOfWeek'].value_counts().idxmax()
print('Busiest day of the week:', maxDOW)
```

```python
# Question 6
"""Is there a correlation between departure delay "DepDelay" and arrival delay "ArrDelay"?
Consider only flights that have a departure delay larger than one and an arrival delay larger than one.  
Calculate the correlation coefficient between these two vectors."""

# filter data so only flights with a departure and arrival delay > 1 are considered
delayedFlights = flights[(flights['DepDelay']>1) & (flights['ArrDelay']>1)]

# calculate correlation between departure delay and arrival delay
corr = delayedFlights['DepDelay'].corr(delayedFlights['ArrDelay'])
print("Correlation between departure and arrival delay:", corr) 
```

```python
# Question 7
"""We will be considering the actual elapsed flight time "ActualElapsedTime" and the travel distance "Distance".
Create a simple linear regression model between these two values for predicting "ActualElapsedTime" using the "Distance" as the explanatory variable. 
Remove flights that do not have an "ActualElapsedTime" or "Distance" value, or their values are NaN or Infinity. """
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# clean the data set to remove rows were actual ecplipsed time or distance do not have a value, are NaN, or are infinity
cleanFlights = flights.dropna(subset=['ActualElapsedTime', 'Distance'])

# find correlation between ActualElapsedTime and Distance
c = cleanFlights['ActualElapsedTime'].corr(cleanFlights['Distance'])

# prepare data for analysis then fit to linear regression model
X = cleanFlights[['Distance']]
y = cleanFlights[['ActualElapsedTime']]
model = LinearRegression()
model.fit(X, y)
slope = model.coef_[0]
intercept = model.intercept_

# predict data from the calculated slope and find how well it fits the data
predictY = model.predict(X)
rSq = r2_score(y, predictY)


print('Correlation between ActualElapsedTime and Distance:', c)
print("Linear Regression Equation: ActualElapsedTime = ", slope, "*Distance + ", intercept, sep='')
print("R-squared for the model:", rSq)
 
```
