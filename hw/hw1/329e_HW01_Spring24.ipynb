## Q2 - What is the mean departure delay of all flights considering only the positive departure delay values? (2 points)
Negative delay value means that the flight left earlier than scheduled. 


```python
delays = df[df["DEPARTURE_DELAY"]>0]
delays["DEPARTURE_DELAY"].mean()
```

# Q3 - Find a list of all airport (Origin or destination airports). (2 points)
The returned value should be a list in Python. 


```python
airports = np.unique(df[["ORIGIN_AIRPORT", "DESTINATION_AIRPORT"]].values)
airports.tolist()
```

# Q4 - Find out which origin airport has the largest delay on Weekends. (2 points)
Weekend days have the IDs of DAY_OF_WEEK 6 and 7. 


```python
wknds = delays[delays["DAY_OF_WEEK"].isin((6, 7))]
origins = wknds.groupby("ORIGIN_AIRPORT")
origins["DEPARTURE_DELAY"].mean().idxmax()
```

# Q5 - Find the top 3 Airlines which had the largest mean departure delay in the month of January (2 points)
Month of January has the id of 1. Consider only flights with the positive departure delay. 

```python
jan_delays = delays[delays["MONTH"] == 1]
avg_jan_delays = jan_delays.groupby("AIRLINE")["DEPARTURE_DELAY"].mean()
avg_jan_delays.nlargest(3, keep="first")
```

# Q7 - Which origin airport has the most cancellation of departure flights? (2 points)

## Q8 - Create a histogram plot that shows the data distribution of the arrival delays consider only positive arrival delays and smaller than 200 minutes. Use 10 minutes bin/bucket size. (2 points)

```python
import matplotlib.pyplot as plt

arrival_delays = df[(df['ARRIVAL_DELAY'] > 0) & (df['ARRIVAL_DELAY'] < 200)]

plt.hist(arrival_delays['ARRIVAL_DELAY'], bins=range(0, 200, 10), edgecolor='black')
plt.title('Distribution of Arrival Delays (0-200 minutes)')
plt.xlabel('Arrival Delay (minutes)')
plt.ylabel('Number of Flights')
plt.xticks(range(0, 201, 10))

plt.show()
```

## Q9 -  Plot a scatter plot of  flight travel air time and arrival delays. Describe if there is any corrolation between them. (2 points)

Describe and interpret the correlation between the two variables of distance and air travel time. 

```python
plt.scatter(df['AIR_TIME'], df['ARRIVAL_DELAY'])
plt.title('Scatter Plot of Flight Air Time vs. Arrival Delay')
plt.xlabel('Air Time (minutes)')
plt.ylabel('Arrival Delay (minutes)')

plt.show()
```

```python
correlation = df['AIR_TIME'].corr(df['ARRIVAL_DELAY'])
correlation
```

The scatter plot of flight air time versus arrival delay reveals that there is a very weak negative correlation (-0.024) between these two variables. This suggests that, generally, the duration of the flight (air time) does not significantly influence the length of the arrival delay.

In other words, longer flights do not consistently lead to longer delays, and shorter flights do not necessarily result in shorter delays. The correlation is so weak that it's almost negligible, indicating that other factors, likely unrelated to air time, play a more significant role in determining arrival delays.

# Question - 10: Which top-3 destination cities has the highst average departure delays on Weekends? (2 points)

* Consider only the positive departure delays. 
* Weekend flights have a day of week of 6 and 7
* You need to merge the airport dataset with the flight dataset to get the destination city of each flight
* You need to group by city and calculate the mean departure delays 
* At the end print out the top 3 of the departure delays with the average delays

```python
# Merging the flights dataset with the airports dataset to get the destination city
merged_df = df.merge(airports, left_on='DESTINATION_AIRPORT', right_on='iata')

# Filtering for weekends and positive departure delays
weekend_flights = merged_df[(merged_df['DAY_OF_WEEK'].isin([6, 7])) & (merged_df['DEPARTURE_DELAY'] > 0)]

# Grouping by destination city and calculating mean departure delays
mean_delays_by_city = weekend_flights.groupby('city')['DEPARTURE_DELAY'].mean()

# Finding the top 3 cities with the highest average departure delays
top_3_cities_delays = mean_delays_by_city.nlargest(3)
top_3_cities_delays
```
