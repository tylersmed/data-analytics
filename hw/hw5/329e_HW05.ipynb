# Task - 1 (5 points)
We want to use logistic regerssion to perdict if the patients will have heart problems or not. The column "Target" in our datasets includes data about heart diseases. If the patient had heart disease we have a 1 and if not a zero. 

Prepare your data set for predicting heart disease ("Target" column) out of 3 features:

* Age of the patient (Column **"Age"**)
* Gender of the patient (male or female - Column **"Sex"**)
* Cholestrol level of the patient (Column **"Chol"**) 


Split your data into 80% traning data and 20% test data. 

```python
from sklearn.model_selection import train_test_split

# split target and predicting variables into training and testing datasets
target_train, target_test, x_multi_train, x_multi_test = train_test_split(heart_df['Target'], heart_df[['Age', 'Sex', 'Chol']], test_size=0.2, random_state=0)

```

# Task 2 - (10 points)

Generate a logistic regression model using your training data. 

Print out the accuracy if your model. 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()

# fit the data to a logistic regressin model
model.fit(x_multi_train, target_train)

# use the logistic regression model to predict the target with testing dataset
# then calculate the accuracy score for the prediction
target_prediction = model.predict(x_multi_test)
accuracy = accuracy_score(target_test, target_prediction)

print("Accuracy score of logistic regression model:", accuracy)

```

# Task 3 - (5 points)


Generate the classification report for your Logistic regresion model and interpret your results regarding precision, recall and f1-score. 


```python
from sklearn.metrics import classification_report

# calculate and print the classification report
cr = classification_report(target_test, target_prediction)
print(cr)

# Interpertation of results:
# Precision: 81% of predicted negatives for heart diseases were correct but only 69% of positives predicted were correct
# Recall: 74% of people without heart disease were predicted correctly and 77% of those with heart disease were predicted correctly
# F1-score: the harmoic mean of precision and recal is 78% for those without heart disease and 73% for those with heard disease
```
